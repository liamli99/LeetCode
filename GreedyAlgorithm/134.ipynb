注意两点！一，总油量减去总消耗如果大于等于0，则必然可以跑完一圈；二，如果从0开始累加gas-cost，在i位置和小于0，则[0, i]区间的任何一个点都不能作为起点！因为他们在位置i处的累和都小于0！证明如下：
我们从i+1处开始重新累加gas-cost！   
局部最优在于每一处节点gas-cost都要大于等于0，这样最终才能跑完一圈！

```python
# 暴力求解，exceed time limit!
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        for index in range(len(gas)):
            sum = 0
            for i in range(index, index + len(gas)):
                sum = sum + gas[i % len(gas)] - cost[i % len(gas)]
                if sum < 0:
                    break
                if i == index + len(gas) - 1:
                    return index
        return -1
```

```python
# 贪心
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        curSum = 0  
        totalSum = 0  
        start = 0 
        
        for i in range(len(gas)):
            curSum += gas[i] - cost[i]
            totalSum += gas[i] - cost[i]
            
            if curSum < 0:  
                start = i + 1  
                curSum = 0  
        
        if totalSum < 0:
            return -1  
        
        return start
```
