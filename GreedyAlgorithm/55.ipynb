# Approach
我们不用考虑当前位置跳几步，只需要考虑当前位置可跳的覆盖范围！确保两点：1. 最大范围一定要覆盖到最后一个元素，这样才能确保能跳到最后一个元素 2. 中间每个元素都要被最大范围覆盖到，这样才能确保可以连续跳跃而不中断！


贪心在于想让覆盖范围尽可能的大，所以我们定义局部最优：每个位置都取最大范围，不断更新最大范围，让最后的范围超过最后一个元素！


# Code

```python
class Solution:
    def canJump(self, nums: List[int]) -> bool:
        max_ = 0

        for i in range(len(nums)):
            if max_ < i:
                return False
            else:
                max_ = max(max_, i + nums[i])
        
        return True
```
