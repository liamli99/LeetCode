贪心在于每一次跳跃尽量多跳！   
每一次跳跃都有覆盖范围！我们需要有两个变量，一个记录当前跳跃的覆盖范围，另一个记录下次跳跃的覆盖范围！遍历当前跳跃的覆盖范围，选择其中可以使得下次跳跃覆盖范围最广的那次跳跃！

# Code

```python
# 贪心，我写的，超时了！
class Solution:
    def jump(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return 0

        curr_max = 0
        next_max = 0
        
        jump = 0

        startI = 0
        maxI = 0
        while True:   
            for i in range(startI, curr_max + 1):
                if i + nums[i] >= len(nums) - 1:
                    jump += 1
                    return jump
                elif i + nums[i] > curr_max:
                    next_max = i + nums[i]
                    maxI = i
                i += 1
            startI = maxI
            curr_max = next_max
            jump += 1
        
        return jump
```

其实根本没必要去找maxI！某次跳跃的起点只需要从上上次跳跃范围终点开始找即可！如下图所示：


```python
# 比我的效率更高！
class Solution:
    def jump(self, nums: List[int]) -> int:
        if len(nums) == 1:
            return 0
            
        curr_max = 0
        next_max = 0 
        jump = 0

        for i in range(len(nums)):
            # 找到curr_max及其之前的最大范围！
            next_max = max(next_max, i + nums[i]) 
            
            if i == curr_max:
                # 开始跳跃一次！
                jump += 1
                curr_max = next_max
                if next_max >= len(nums) - 1:
                    break

        return jump
```
