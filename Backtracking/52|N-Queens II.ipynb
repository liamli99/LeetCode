# Note
回溯算法模板框架：
```
void backtracking(参数) {
    if (终止条件) {
        存放结果;
        return;
    }

    for (选择：本层集合中元素（树中节点孩子的数量就是集合的大小）) {
        处理节点;
        backtracking(路径，选择列表); // 递归
        回溯，撤销处理结果
    }
}
```
本题约束条件：1.不能同行 2.不能同列 3.不能同斜线


# Code

````python
# result函数存储数量，现在不需要所有结果了！！！
class Solution:
    def totalNQueens(self, n: int) -> int:
        path = []
        # 注意这里不能写成result=0！！！如果这样写就说明result是整数，被当作参数传递到backtracking函数中时，在backtracking内部
        # 对result的累加不会反应到backtracking外部的result！即最后返回的一直都是0！这是因为整数是immutable！！！
        result = [0]
        self.backtracking(n, 0, path, result)
        return result[0]

    def backtracking(self, n, row, path, result):
        if row == n:
            result[0] += 1
            return
        
        for col in range(n):
            if self.isValid(n, row, col, path):
                l = ["."] * n
                l[col] = "Q"
                path.append(l)
                self.backtracking(n, row + 1, path, result)
                path.pop()

    def isValid(self, n, row, col, path):
        for i in range(row):
            if path[i][col] == "Q":
                return False
        
        i, j = row - 1, col - 1
        while i >= 0 and j >= 0:
            if path[i][j] == "Q":
                return False
            i -= 1
            j -= 1

        i, j = row - 1, col + 1
        while i >= 0 and j < n:
            if path[i][j] == "Q":
                return False
            i -= 1
            j += 1
        
        return True
````
