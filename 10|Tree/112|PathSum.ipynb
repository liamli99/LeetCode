```python
# preorder, recursion
# 方法一
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if root == None:
            return False
       
        path = []
        return self.traversal(root, path, targetSum)
    
    def traversal(self, node, path, targetSum):
        path.append(node.val)
        
        if node.left == None and node.right == None and sum(path) == targetSum:  
            return True
        
        if node.left != None:
            if self.traversal(node.left, path, targetSum):
                return True
            # 回溯，由于path是list，list是mutable，因此path在递归中会发生改变！我们要将加进path的node.left.val移除！
            path.pop()
     
        if node.right != None:
            if self.traversal(node.right, path, targetSum):
                return True
            path.pop()
        
        return False
```

```python
# preorder, recursion
# 方法二
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if root == None:
            return False
        
        return self.traversal(root, targetSum - root.val)
        
    def traversal(self, node, sum_):
        if node.left == None and node.right == None and sum_ == 0:
            return True
        
        # 注意下方的左和右分别代表两种回溯的写法！一种是明显的回溯，另一种是隐藏的回溯，两者都正确！
        
        # 左
        if node.left != None:
            sum_ -= node.left.val
            if self.traversal(node.left, sum_):
                return True
            # 回溯，由于sum_是integer，integer是immutable，因此sum_在递归中不会发生改变！但是递归前sum_减去了node.left.val，
            # 因此这里的回溯是加上node.left.val！
            sum_ += node.left.val
        
        # 右
        if node.right != None:
            if self.traversal(node.right, sum_ - node.right.val):
                return True
            # 不需要回溯！由于sum_是integer，integer是immutable，因此sum_在递归中不会发生改变！而且递归前也没有改变sum_，因此不需要回溯！
            # 仔细体会这里左和右的写法，理解为什么二者是等价的！
        
        return False
```

```python
# preorder, recursion
# 方法三
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if root == None:
            return False
        
        return self.traversal(root, targetSum)
        
    def traversal(self, node, sum_):
        # 中
        sum_ -= node.val
        
        if node.left == None and node.right == None and sum_ == 0:
            return True
        
        # 左
        if node.left != None:
            if self.traversal(node.left, sum_):
                return True
            # 不需要回溯！由于sum_是integer，integer是immutable，因此path在递归中不会发生改变！
        
        # 右
        if node.right != None:
            if self.traversal(node.right, sum_):
                return True
        
        return False
```

```python
# preorder, iteration
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if root == None:
            return False
        
        stack = [root]
        sum_lst = [root.val]
        
        while len(stack) != 0:
            node = stack.pop()
            sum_ = sum_lst.pop()
            
            if node.left == None and node.right == None and sum_ == targetSum:
                return True

            if node.right != None:
                stack.append(node.right)
                sum_lst.append(sum_ + node.right.val)
            
            if node.left != None:
                stack.append(node.left)
                sum_lst.append(sum_ + node.left.val)
    
        return False    
```
