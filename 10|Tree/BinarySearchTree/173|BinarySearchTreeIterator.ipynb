# Approach   
本题就是要实现一个基于中序遍历的iterator，即将中序遍历的代码拆解成_init_，next，hasNext！   

# Note
Inorder Traversal (左 中 右)
```python
# iteration 1
def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if root == None:
            return []
        
        stack = []  
        curr = root
        result = []
        
        while curr != None or len(stack) != 0:
            if curr != None:     
                stack.append(curr)
                curr = curr.left
            else:
                curr = stack.pop()
                result.append(curr.val)
                curr = curr.right
        
        return result

# iteration 2
def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        if root == None:
            return []
        
        stack = []
        curr = root
        result = []

        while curr != None or len(stack) != 0:
            while curr != None:
                stack.append(curr)
                curr = curr.left
            curr = stack.pop()
            result.append(curr.val)
            curr = curr.right
        
        return result
```
以上两种iteration的写法完全等价！   

# Code

````python
class BSTIterator:

    def __init__(self, root: Optional[TreeNode]):
        self.stack = []  
        self.curr = root
        
        while self.curr != None:
            self.stack.append(self.curr)
            self.curr = self.curr.left

    def next(self) -> int:
        self.curr = self.stack.pop()
        result = self.curr.val
        self.curr = self.curr.right
        
        while self.curr != None:
            self.stack.append(self.curr)
            self.curr = self.curr.left
        
        return result

    def hasNext(self) -> bool:
        return len(self.stack) != 0
````
