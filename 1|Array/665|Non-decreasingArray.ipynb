# Approach
注意题目中说我们最多只能修改一个数字，因此nums中最多只能存在一个decreasing的数字(该数字比其左边的数字要小)！因此我们需要一个变量count去记录nums中decreasing的数字的数量，如果count等于2，就返回False！   

当我们第一次遇到decreasing的数字nums[i]时，有两种修改方案：1.将nums[i-1]变成nums[i]。如果nums[i-2]不存在的话，即i==1时，必然可以这么修改！如果nums[i-2]存在的话，此时需要判断nums[i-2]是否小于等于nums[i]，只有确定小于等于才可以这样修改！2.将nums[i]变成nums[i-1]。如果nums[i+1]不存在的话，即i==len(nums)-1时，必然可以这么修改！如果nums[i+1]存在的话，此时需要判断nums[i+1]是否大于等于nums[i-1]，只有确定大于等于才可以这样修改！如果以上两种方案都不行，那就返回False！

# Code

```python
class Solution:
    def checkPossibility(self, nums: List[int]) -> bool:
        # 记录decreasing number的数量
        count = 0
        for i in range(1, len(nums)):
            if nums[i] < nums[i - 1]:
                count += 1
                
                if count == 2:
                    return False
                
                # a = nums[i - 2] if i > 1 else float('-inf')
                # b = nums[i + 1] if i < len(nums) - 1 else float('inf')
                # if not (a <= nums[i] or b >= nums[i - 1]):
                if not ((i == 1 or nums[i - 2] <= nums[i]) or (i == len(nums) - 1 or nums[i + 1] >= nums[i - 1])):
                    return False
        
        return True
```
