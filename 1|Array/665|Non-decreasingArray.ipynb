```python
class Solution:
    def checkPossibility(self, nums: List[int]) -> bool:
        # decreasing number
        count = 0
        for i in range(1, len(nums)):
            if nums[i] < nums[i - 1]:
                count += 1
                
                if count == 2:
                    return False
                
                # a = nums[i - 2] if i > 1 else float('-inf')
                # b = nums[i + 1] if i < len(nums) - 1 else float('inf')
                # if not (a <= nums[i] or b >= nums[i - 1]):
                if not ((i == 1 or nums[i - 2] <= nums[i]) or (i == len(nums) - 1 or nums[i + 1] >= nums[i - 1])):
                    return False
        
        return True
```
