# Approach
将本题抽象成树形结构：


```python
# Time: O(3^m + 4^n), Space: O(m + n)，m和n分别是digits中对应3个字母的数字个数和对应4个字母的数字个数
from typing import List
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        # 如果没有这行，那么当digits为''时，返回的是['']
        if digits == '':
            return []
        
        digit_to_letter = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz'
        }

        path = ''
        result = []
        # 注意本题的startIndex不同于77的startIndex，本题的startIndex用来记录遍历到第几个digit，即此时digits中的位置，也表示树的深度
        startIndex = 0
        self.backtracking(digits, digit_to_letter, startIndex, path, result)
        return result
    
    def backtracking(self, digits, digit_to_letter, startIndex, path, result):
        if len(path) == len(digits):
            result.append(path)
            return
        
        digit = digits[startIndex]
        letters = digit_to_letter[digit]

        for letter in letters:
            path += letter
            self.backtracking(digits, digit_to_letter, startIndex + 1, path, result)
            path = path[:-1]
```
