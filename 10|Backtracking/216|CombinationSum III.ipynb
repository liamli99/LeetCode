剪枝优化：      
1.如果path中还没到k个元素，此时总和已经等于n了，也得终止！   
2.类同77，假如此时path里面已经有len(path)个元素，我们要k个，因此还需要取k-len(path)个元素！由于最多也就只能取到9，因此我们最晚也要从9-(k-len(path))+1开始取！因为9-(k-len(path))+1到9之间正好有k-len(path)个元素！因此我们的i最多也就只能取到9-(k-len(path))+1！

# Note
回溯算法模板框架：
void backtracking(参数) {
    if (终止条件) {
        存放结果;
        return;
    }

    for (选择：本层集合中元素（树中节点孩子的数量就是集合的大小）) {
        处理节点;
        backtracking(路径，选择列表); // 递归
        回溯，撤销处理结果
    }
}

# Code

```python
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        path = []
        result = []
        self.backtracking(k, n, 1, path, result)
        return result
    
    def backtracking(self, k, n, startIndex, path, result):
        # 两条限制：和为n且只能有k个元素！
        # 不要写成len(path)==k and sum(path)==n，这样的话如果已经有k个元素但是sum还不是n的话，path还会继续添加元素的！
        # 但是哪怕这样写了也不用怕无限递归，因为每次递归的下一层就是从i+1开始的，有限制！
        if len(path) == k:
            if sum(path) == n:
                result.append(path[:])
            return
        # 题目要求只能从1到9中选取数字！由于每个数字只能用一次而且答案不能重复，还要用startIndex！
        for i in range(startIndex, 10):
            path.append(i)
            self.backtracking(k, n, i + 1, path, result)
            path.pop()

# 如果不能用sum函数求和，只能定义一个参数去累加！
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        path = []
        result = []
        self.backtracking(k, n, 1, 0, path, result)
        return result
    
    def backtracking(self, k, n, startIndex, sum, path, result):
        if len(path) == k and sum == n:
            result.append(path[:])
            return
        
        for i in range(startIndex, 10):
            path.append(i)
            sum += i
            self.backtracking(k, n, i + 1, sum, path, result)
            sum -= i
            path.pop()
```

```python
# 剪枝优化
class Solution:
    def combinationSum3(self, k: int, n: int) -> List[List[int]]:
        path = []
        result = []
        self.backtracking(k, n, 1, path, result)
        return result
    
    def backtracking(self, k, n, startIndex, path, result):
        # 优化1
        if sum(path) > n:
            return
        
        if len(path) == k:
            if sum(path) == n:
                result.append(path[:])
            return
        
        # 优化2
        for i in range(startIndex, 9 - (k - len(path)) + 2):
            path.append(i)
            self.backtracking(k, n, i + 1, path, result)
            path.pop()
```
