# Approach
- 首先利用141判断是否有环，若有环，再找入口！    

- 找环的入口：根据141，如果链表有环，则快指针必定能在环中与慢指针相遇(套圈)！假设head到环入口的节点数为x，环入口到相遇点的节点数为y，相遇点再到环入口的节点数为z：


- 相遇时慢指针还没走完一圈环，但是快指针至少已经走完一圈环了！因此相遇时慢指针走过的节点数为x+y，快指针走过的节点数为x+n(y+z)+y，n为快指针在环内走的圈数，n>=1！两个指针相遇时所花时间相同，而时间等于路程除以速度，因此(x+y)/1=(x+n(y+z)+y)/2，化简得x+y=n(y+z)！由于我们想找环的入口，因此我们更关心x的大小！x=n(y+z)-y，继续化简得x=(n-1)(y+z)+z！

- 这个公式表明，如果有两个指针，一个指针p1从head处出发，每次走一个节点，另一个指针p2从相遇点出发，每次也走一个节点，那么这两个指针必然会在环入口相遇！证明：当p1走到环入口时，用时为x/1=x！那么当p2走了x时间后，走的节点数为x·1=x=(n-1)(y+z)+z，p2也在环入口！因此p1和p2相遇的节点就是环入口！！！   

# Code

```python
# Time: O(n), Space: O(1)

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow, fast = head, head
 
        while fast != None and fast.next != None:
            slow = slow.next
            fast = fast.next.next

            if slow == fast:
                p1, p2 = head, slow
                while p1 != p2:
                    p1 = p1.next
                    p2 = p2.next
                return p1
        
        return None     
```
