# Approach
- 类似77和39，由于本题每个元素只能取一次，因此下一层从i+1开始！

- 注意和77不同，本题candidates中有重复元素！！！因此就算定义了startIndex，也没有办法确保最后的结果中的集合不重复！！！例如candidates = `[10, 1, 2, 7, 6, 1, 5]`，target = 8，如果从1开始，我们可以找到`[1, 2, 5]`；如果从2开始，我们可以找到`[2, 1, 5]`，重复！！！

- 因此，只用startIndex还不够去重！我们还需要去重的是**同一层上重复的元素**！！！
  - 方法一
  将candidates从小到大排序，在当前层中，如果`i > startIndex`并且`candidates[i] == candidates[i - 1]`，说明`candidates[i]`重复了，直接跳过！

  - 方法二
  将candidates从小到大排序，构建一个used数组，`used[i]`表示`candidates[i]`是否在path中！在当前层中，如果`i > startIndex`并且`candidates[i] == candidates[i - 1]`并且`used[i - 1] == False`，说明`candidates[i]`重复了，直接跳过！注意一定要强调`used[i - 1] == False`，因为只有`candidates[i - 1]`不在path中，才能证明`candidates[i - 1]`和`candidates[i]`是同一层的元素！！！

  注意，方法二利用used数组来去掉本层的重复元素，这才是普适的方法！！！而本题之所以能用方法一，是因为定义了startIndex，从而可以确保`candidates[i - 1]`和`candidates[i]`在同一层！！！

- 剪枝操作和39完全相同！


# Code

```python
# Time: O(n * 2^n), Space: O(target)
# 方法一
from typing import List
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        path = []
        result = []
        # 为了去重和剪枝
        candidates.sort()

        def backtrack(startIndex, currSum):
            if currSum == target:
                result.append(path[:])
                return
            
            for i in range(startIndex, len(candidates)):
                # 去重
                if i > startIndex and candidates[i] == candidates[i - 1]:
                    continue
                
                # 剪枝
                if currSum + candidates[i] > target:
                    break

                path.append(candidates[i])
                currSum += candidates[i]
                backtrack(i + 1, currSum)
                path.pop()
                currSum -= candidates[i]
            
        backtrack(0, 0)
        return result
    
```

```python
# Time: O(n * 2^n), Space: O(target)
# 方法二
class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        path = []
        result = []
        # 为了去掉本层的重复元素
        used = [False] * len(candidates)
        # 为了去重和剪枝
        candidates.sort()

        def backtrack(startIndex, currSum):
            if currSum == target:
                result.append(path[:])
                return
            
            for i in range(startIndex, len(candidates)):
                # 去重
                if i > startIndex and candidates[i] == candidates[i - 1] and used[i - 1] == False:
                    continue
                
                # 剪枝
                if currSum + candidates[i] > target:
                    break

                path.append(candidates[i])
                currSum += candidates[i]
                backtrack(i + 1, currSum)
                path.pop()
                currSum -= candidates[i]
            
        backtrack(0, 0)
        return result
    
```
