```python
# Time: O(3^m + 4^n), Space: O(m + n)mndigits34
from typing import List
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        # digits''['']
        if digits == '':
            return []
        
        digit_to_letter = {
            '2': 'abc',
            '3': 'def',
            '4': 'ghi',
            '5': 'jkl',
            '6': 'mno',
            '7': 'pqrs',
            '8': 'tuv',
            '9': 'wxyz'
        }

        path = ''
        result = []
        # startIndex77startIndexstartIndexdigitdigits
        startIndex = 0
        self.backtracking(digits, digit_to_letter, startIndex, path, result)
        return result
    
    def backtracking(self, digits, digit_to_letter, startIndex, path, result):
        if len(path) == len(digits):
            result.append(path)
            return
        
        digit = digits[startIndex]
        letters = digit_to_letter[digit]

        for letter in letters:
            path += letter
            self.backtracking(digits, digit_to_letter, startIndex + 1, path, result)
            path = path[:-1]
```
