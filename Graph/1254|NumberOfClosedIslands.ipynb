# Approach  
本题类同1020和130！   

本题的Closed Islands就是1020中的Enclaves组成的区域和130中的surrounded region！！！   

具体方法：先遍历grid的四个边界的格子，采用200中的DFS/BFS方法将所有相邻的且值为0的格子都进行标记！此时剩下的未被标记且值为0的格子组成的区域就是closed islands！我们需要再次遍历grid中的所有格子，利用200的方法计算未被标记且值为0的格子可以组成islands的数量，这就是closed islands的数量！！！   

本题的方法和130与1020基本相同，唯一的明显区别就是本题标记完之后再次遍历时要用200的方法计算islands的数量，而130标记完之后再次遍历时遇到未标记的O就将其变成X，1020标记完之后再次遍历时遇到未标记的1就计数加1！   

化简：上述方法可以化简成只需要使用一次nested for loop，我们需要修改DFS/BFS！此时的DFS/BFS不仅标记相邻的且值为0的格子(island)，而且会有一个返回值！返回值为0表示此时标记的island不是closed island，返回值为1表示此时标记的island时closed island！因此可以在遍历grid的所有格子的过程中直接累加未被标记且值为0的格子DFS/BFS之后的结果！注意仔细理解DFS/BFS的代码，非常巧妙！

```python
# DFS
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        used = [[False] * len(grid[0]) for _ in range(len(grid))]
        sum_ = 0
        
        # 四个边界
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if used[i][j] == False and grid[i][j] == 0 and (i == 0 or i == len(grid) - 1 or j == 0 or j == len(grid[0]) - 1):
                    self.dfs(grid, i, j, used)
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if used[i][j] == False and grid[i][j] == 0:
                    sum_ += 1
                    self.dfs(grid, i, j, used)
        
        return sum_
    
    def dfs(self, grid, row, col, used):
        if (row < 0 or row >= len(grid) or col < 0 or col >= len(grid[0]) or used[row][col] == True or grid[row][col] == 1):
            return 
        
        used[row][col] = True
        
        self.dfs(grid, row - 1, col, used)
        self.dfs(grid, row + 1, col, used) 
        self.dfs(grid, row, col - 1, used)
        self.dfs(grid, row, col + 1, used)

# 化简
class Solution:
    def closedIsland(self, grid: List[List[int]]) -> int:
        used = [[False] * len(grid[0]) for _ in range(len(grid))]
        sum_ = 0
        
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if used[i][j] == False and grid[i][j] == 0:
                    sum_ += self.dfs(grid, i, j, used)
        
        return sum_
    
    def dfs(self, grid, row, col, used):
        if row < 0 or row >= len(grid) or col < 0 or col >= len(grid[0]):
            return 0
        
        if used[row][col] == True or grid[row][col] == 1:
            return 1
        
        used[row][col] = True
        
        return min(self.dfs(grid, row - 1, col, used),\
                   self.dfs(grid, row + 1, col, used),\
                   self.dfs(grid, row, col - 1, used),\
                   self.dfs(grid, row, col + 1, used))
```
