# Approach
本题是79的拓展，79只需要判断一个单词是否在board中，但是本题要判断很多单词！！！   

79的方法是遍历board中的每个字母，对每个字母都用backtracking来判断是否存在以该字母开头的单词！仿照79，本题也要遍历board中的每个字母，对每个字母都用backtracking来判断是否存在以该字母开头的**所有单词**，即要同时检查words中的所有单词！仔细体会79和212的代码！      

我们可以把words中的每个单词都存在Trie中！79的逻辑是将遍历到的某个board[row][col]和word[0]比较，如果相等，就将board[row][col]上下左右四个方向的相邻字母与word[1]比较，重复上述操作！本题的逻辑是判断遍历到的某个board[row][col]是否在root.children中，如果在，就判断board[row][col]上下左右四个方向的相邻字母是否在root.children[board[row][col]].children中，重复上述操作！这样就相当于同时检查words中的所有单词！    

# Note
Trie(字典树或前缀树)，根节点是空字符，其余每个节点的值都是一个字符！从根节点的孩子一直到某个节点这个路径上的所有字符可以组成一个字符串，这就是该节点所对应的字符串！因此一个节点的所有子节点所对应的字符串都有相同的前缀，即该节点所对应的字符串！ 

# Code

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isEndOfWord = False
    
    def insert(self, word):
        curr = self
        
        for char in word:
            if char not in curr.children:
                curr.children[char] = TrieNode()
            
            curr = curr.children[char]
        
        curr.isEndOfWord = True

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        root = TrieNode()
        for word in words:
            root.insert(word)
        
        path = ""
        result = set()
        
        used = [[False] * len(board[0]) for _ in range(len(board))]

        for i in range(len(board)):
            for j in range(len(board[0])):
                self.backtracking(board, i, j, used, root, path, result)

        return list(result)
    
    # node就相当于79的index！
    def backtracking(self, board, row, col, used, node, path, result):
        if (row < 0 or row >= len(board) or col < 0 or col >= len(board[0]) or used[row][col] or board[row][col] not in node.children):
            return 
        
        used[row][col] = True
        
        node = node.children[board[row][col]]
        path += board[row][col]
        if node.isEndOfWord == True:
            result.add(path)
        
        self.backtracking(board, row - 1, col, used, node, path, result)
        self.backtracking(board, row + 1, col, used, node, path, result)
        self.backtracking(board, row, col - 1, used, node, path, result)
        self.backtracking(board, row, col + 1, used, node, path, result)
 
        used[row][col] = False
```

```python
# 让上述代码更加清晰，根据208，insert在Trie class中更合理！
class TrieNode:
    def __init__(self):
        self.children = {}
        self.isEndOfWord = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        curr = self.root
        
        for char in word:
            if char not in curr.children:
                curr.children[char] = TrieNode()
            curr = curr.children[char]
        
        curr.isEndOfWord = True

class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        trie = Trie()
        for word in words:
            trie.insert(word)
        root = trie.root
        
        path = ""
        result = set()
        
        used = [[False] * len(board[0]) for _ in range(len(board))]

        for i in range(len(board)):
            for j in range(len(board[0])):
                
                self.backtracking(board, i, j, used, root, path, result)

        return list(result)

    def backtracking(self, board, row, col, used, node, path, result):
        if (row < 0 or row >= len(board) or col < 0 or col >= len(board[0]) or used[row][col] or board[row][col] not in node.children):
            return 
        
        used[row][col] = True
        
        node = node.children[board[row][col]]
        path += board[row][col]
        if node.isEndOfWord == True:
            result.add(path)
        
        self.backtracking(board, row - 1, col, used, node, path, result)
        self.backtracking(board, row + 1, col, used, node, path, result)
        self.backtracking(board, row, col - 1, used, node, path, result)
        self.backtracking(board, row, col + 1, used, node, path, result)
 
        used[row][col] = False
```
