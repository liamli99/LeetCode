# Approach   
本题可以用recursion和iteration两个方法来写！   

在recursion中，我们只需要具体写root不是None且root.val==key时的代码，即删除的节点是root！   

在iteration中，我们要用curr和pre两个指针找到要删除的节点和其父节点！如果curr是root，那么方法类同recursion！如果curr不是root，那么先用recursion中的方法把curr删掉得到一个新的子树，再让pre指向这个新的子树！   


# Note
删除root有很多种方法：1.找root右子树的最小值，即右子树中最左下方的节点，将该最小值赋给root，然后再在右子树中调用递归函数来删除这个值最小的节点 2.找到root左子树的最大值，即左子树中最右下方的节点，将该最大值赋给root，然后再在左子树中调用递归函数来删除这个值最大的节点 3.找到root右子树的最小值，即右子树中最左下方的节点，让root的左子树成为这个节点的左子树，然后将root替换成root的右子树 4.找到root左子树的最小值，即左子树中最右下方的节点，让root的右子树成为这个节点的左子树，然后将root替换成root的左子树

```python
# recursion
class Solution:
    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        if root == None:
            return None
        
        if key < root.val:
            root.left = self.deleteNode(root.left, key)
        elif key > root.val:
            root.right = self.deleteNode(root.right, key)
        else:
            if root.left == None:
                return root.right
            if root.right == None:
                return root.left
            
            # 方法一
            curr = root.right
            while curr.left != None:
                curr = curr.left 
            root.val = curr.val
            root.right = self.deleteNode(root.right, curr.val)
            
            # 方法二
            # curr = root.left
            # while curr.right != None:
            #     curr = curr.right 
            # root.val = curr.val
            # root.left = self.deleteNode(root.left, curr.val)
            
            # 方法三
            # curr = root.right
            # while curr.left != None:
            #     curr = curr.left 
            # curr.left = root.left
            # root = root.right
            
            # 方法四
            # curr = root.left
            # while curr.right != None:
            #     curr = curr.right 
            # curr.right = root.right
            # root = root.left
 
        return root
```

```python
# iteration
class Solution:
    def deleteNode(self, root: Optional[TreeNode], key: int) -> Optional[TreeNode]:
        if root == None:
            return None
        
        # 用curr和pre来找到值为key的节点及其父节点！
        pre, curr = None, root
        while curr != None:
            if curr.val == key:
                break
            pre = curr
            if key < curr.val:
                curr = curr.left
            elif key > curr.val:
                curr = curr.right
        
        # 如果要删除的节点就是root，那么方法完全类同iteration，直接调用函数！
        if curr == root:
            return self.delete(curr)
        # 如果要删除的节点不是root且存在，先调用函数将该节点删掉得到一个新的子树，再让pre指向这个新的子树！
        elif curr != None:
            if curr.val < pre.val:
                pre.left = self.delete(curr)
            elif curr.val > pre.val:
                pre.right = self.delete(curr)
        
        return root  
    
    # 删除root的方法三！
    def delete(self, node):
        if node == None:
            return None
        
        if node.left == None:
            return node.right
        if node.right == None:
            return node.left
        
        curr = node.right
        while curr.left != None:
            curr = curr.left 
        curr.left = node.left
        return node.right
```
